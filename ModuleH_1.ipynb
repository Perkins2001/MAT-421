## 22.1 ODE Initial Value Problem Statement

An ODE involves a function \( y(x) \), its independent variable \( x \), and its derivatives. Consider a second-order ODE given by:
$$ y'' + p(x)y' + q(x)y = r(x) $$
where \( p(x) \), \( q(x) \), and \( r(x) \) are known functions, and \( y \) is the function we want to find. The initial conditions might be given at \( x_0 \) as \( y(x_0) = y_0 \) and \( y'(x_0) = y'_0 \).

## 22.2 Reduction of Order

To solve higher-order ODEs, we may need to reduce the order. For instance, if we have a second-order ODE like \( y'' - xy' + 4y = 0 \), we can set a new variable \( v = y' \) to reduce it to a system of first-order ODEs.


## 22.3 The Euler Method

The Euler method can be used to numerically solve a first-order ODE. This method iteratively computes the value of the function at discrete points by approximating the tangent line at each point.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ODE and step size
f = lambda x, y: x - y  # Example ODE
h = 0.05  # Step size
x = np.arange(0, 2 + h, h)  # Numerical grid
y0 = 1  # Initial condition

# Initialize the array for y values
y = np.zeros(len(x))
y[0] = y0

# Euler's Method to solve the ODE
for i in range(1, len(x)):
    y[i] = y[i - 1] + h * f(x[i - 1], y[i - 1])

# Plotting the results
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'b--', label='Approximate')
plt.title('Approximate Solution using Euler’s Method')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid(True)
plt.legend()
plt.show()
```
