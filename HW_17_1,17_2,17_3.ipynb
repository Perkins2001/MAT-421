Interpolation works well when the model is completley linear, the example below uses the function y=4x.

```python
import numpy as np
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [4,8,12,16,20]

x_new = 3.5
y_new = np.interp(x_new, x, y)
print("The interpolation value is: ",y_new)
print("4 * 3.5 = ", 4*3.5)
```

```python
plt.plot(x, y, 'o-')
plt.title('Interpolation at x = 3.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Now that weve shown linear interpolation I will show an example of Cubic Spline Interpolation.
This model is generally more accurate and smoother by using a higher degree polynomial.

```python
from scipy.interpolate import CubicSpline
x = [0, 1, 2, 3]
y = [0, 3, 0, 4]

f = CubicSpline(x, y, bc_type = 'natural')
x_new = np.linspace(0, 3, 100)
y_new = f(x_new)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
