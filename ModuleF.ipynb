**Numerical Grid:** Evenly spaced set of points over the domain of a function.

**Spacing or step size:** Distance between adjacent points on the numerical grid.

One dimension numerical grids can be generated in python as follows:

```python
import numpy as np
x=np.linspace(1,2, num=5)
print(x)
```

This is a numerical grid with a step size of .25

The derivative fâ€²(x) of a function f(x) at the point x=a is defined as:

The derivative of the function f at the point a is determined by the limit as x approaches a of the expression f(x) - f(a) divided by x - a.

Example of numerical differentiation on the function cos(x)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return np.cos(x)

# Generate numerical grid
x_values = np.linspace(0, 2*np.pi, 100)

# Calculate the values of the function
y_values = f(x_values)

# Use np.diff to compute finite differences
finite_diff = np.diff(y_values) / np.diff(x_values)

# Plot
plt.plot(x_values[:-1], finite_diff, '--', label='Numerical Derivative (using np.diff)')
plt.plot(x_values, -np.sin(x_values), label='Exact Derivative (-sin(x))')
plt.legend()

# Labels
plt.title('Numerical Differentiation Example')
plt.xlabel('x')
plt.ylabel('Derivative')

# Show the plot
plt.grid(True)
plt.show()

```

**Numerical Differntation with Noise**

**Noise:** When value is off by a small amount from what it would be if it were computed from a pure mathematical function

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 2*np.pi, 0.01)
omega = 150
epsilon = 0.2

# Generate a clean sine wave, add noise
y = np.sin(x)
y_noise = y + epsilon * np.cos(omega * x)

# Plot the clean sine wave, the sine wave with noise
plt.figure(figsize=(12, 8))
plt.plot(x, y_noise, 'r-', label='sin(x) + noise')
plt.plot(x, y, 'b-', label='sin(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
